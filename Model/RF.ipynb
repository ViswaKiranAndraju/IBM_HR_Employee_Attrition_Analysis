```python
df.drop(['EmployeeCount','EmployeeNumber','Age','JobLevel'], axis=1)
```

```python
df.info()
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1,3,6,10,14,16,20,21])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, criterion='entropy', random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred,y_test)
print(cm)
accuracy_score(y_pred,y_test)
```
